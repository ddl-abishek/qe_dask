```python
import pandas as pd
```

```python
data_2013_14_chunks = pd.read_csv('./Parking_Violations_Issued_-_Fiscal_Year_2014__August_2013___June_2014_.csv',
                                 chunksize=10000)
```

```python
for chunk in data_2013_14_chunks:
    chunk.to_csv("./dataset_subsets/2013-14_first_10000.csv")
    break
```

```python
column_set_dict = {}

for column in chunk:
    column_set_dict[column] = set()
    for value in chunk[column]:
        column_set_dict[column].add(value)
```

```python
for column in column_set_dict:
    print(column,' : ',len(column_set_dict[column]))
```
