# Reading csv in pandas.io.parsers.TextFileReader

```python
data_2013_14_chunks = pd.read_csv('./Parking_Violations_Issued_-_Fiscal_Year_2014__August_2013___June_2014_.csv',
                                 chunksize=10000)
```

## Analyzing different fields to discern the fields to omit
#### NOTE: len(set()) does not give a valid result on fields with NaN values and counts every instance of NaN as a new value


```python
for field_name in df:
    print(field_name," ; ",len(set(df[field_name]))," unique values")
    print(df[field_name].value_counts())
    print("-----------------------------")
```
